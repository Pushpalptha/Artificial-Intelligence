```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Original data
data = {
    "Date": ["2001-Dec 14-16", "2001-Dec 6-9", "2001-Nov 26-27", 
             "2001-Nov 8-11", "2001-Nov 2-4", "2001-Oct 19-21"],
    "apporove": [86, 86, 87, 87, 87, 88],
    "disapprove": [11, 10, 8, 9, 9, 9],
    "dontknow": [3, 4, 5, 4, 4, 3]
}

# Create DataFrame
marks = pd.DataFrame(data)

# Fix typo in column name
marks.rename(columns={"apporove": "approve"}, inplace=True)

# Extract year-month from Date and convert to datetime
marks["Date_clean"] = pd.to_datetime(marks["Date"].str.extract(r'(\d{4}-[A-Za-z]+)')[0], format='%Y-%b')

# Convert datetime to ordinal for regression
marks["Date_ordinal"] = marks["Date_clean"].apply(lambda x: x.toordinal())

# Create regression plot
sns.regplot(x="Date_ordinal", y="approve", data=marks, marker="o", color="blue", line_kws={"color": "red"})

# Format x-axis ticks to readable dates
plt.xticks(ticks=marks["Date_ordinal"], labels=marks["Date_clean"].dt.strftime('%b-%Y'), rotation=45)

# Customize plot
plt.title("Approval Trend Over Time (Regression)")
plt.xlabel("Date")
plt.ylabel("Approval (%)")
plt.ylim(80, 90)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Original data
data = {
    "Date": ["2001-Dec 14-16", "2001-Dec 6-9", "2001-Nov 26-27", 
             "2001-Nov 8-11", "2001-Nov 2-4", "2001-Oct 19-21"],
    "apporove": [86, 86, 87, 87, 87, 88],
    "disapprove": [11, 10, 8, 9, 9, 9],
    "dontknow": [3, 4, 5, 4, 4, 3]
}

# Create DataFrame
marks = pd.DataFrame(data)

# Fix typo
marks.rename(columns={"apporove": "approve"}, inplace=True)

# Plot histogram and KDE
sns.histplot(marks["approve"], kde=True, bins=5, color="skyblue", edgecolor="black")

# Customize
plt.title("Distribution of Approval Ratings")
plt.xlabel("Approval (%)")
plt.ylabel("Frequency")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Stacked area plot
days=[1,2,3,4,5]

sleeping=[7,8,6,11,7]
eating=[2,3,4,3,2]
working=[7,8,7,2,2]
playing=[8,5,7,8,13]

plt.plot([],[],color='m',label='Sleeping',linewidth=5)           # means you're not plotting any x or y data.
plt.plot([],[],color='c',label='Eating',linewidth=5)
plt.plot([],[],color='r',label='Working',linewidth=5)
plt.plot([],[],color='k',label='Playing',linewidth=5)

plt.stackplot(days,sleeping,eating,working,playing,colors=['m','c','r','k'])

plt.xlabel('x')
plt.ylabel('y')
plt.title('Stck plot')
plt.legend()
plt.show()
```

```python
# Pie Chart
slices=[7,2,2,13]
activites=['Sleeping','Eating','Working','Playing']
cols=['c','m','r','b']

plt.pie(slices,                                  #  Slices:-Revelent sizes of each part 
        labels=activites,
        colors=cols,
        startangle=90,
        shadow=True,
        explode=(0,0.1,0,0),                      # to pull out a slice a bit eating is pulled out first. If want to pull out the first slice a bit we would do(0.1,0,0,0)
        autopct='%1.1f%%')                        # overlay the percentages  on the graph itself.

plt.title('PiePlot')
plt.show()
```

```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)

plt.subplot(211)                                    # Subplot:- It helps us to plot multiple plots
plt.plot(t1, f(t1),'bo',t2,f(t2))

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2))
plt.show()
```
