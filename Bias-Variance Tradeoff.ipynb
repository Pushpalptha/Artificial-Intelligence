```python
import matplotlib.pyplot as plt
import numpy as np

# Data
np.random.seed(0)
X = np.linspace(-3, 3, 30)
y = np.sin(X) + np.random.normal(0, 0.2, X.shape)

# Curve
X_curve = np.linspace(-3, 3, 200)
y_true = np.sin(X_curve)

# Underfit (Linear)
y_underfit = np.poly1d(np.polyfit(X, y, 1))(X_curve)

# Plot
plt.scatter(X, y, color="black", label="Data")
plt.plot(X_curve, y_true, "g--", label="True function")
plt.plot(X_curve, y_underfit, "r", label="Underfit")
plt.title("Underfitting (High Bias)")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
np.random.seed(0)
X = np.linspace(-3, 3, 30)
y = np.sin(X) + np.random.normal(0, 0.2, X.shape)

# Curve
X_curve = np.linspace(-3, 3, 200)
y_true = np.sin(X_curve)

# Overfit (Polynomial degree 15)
y_overfit = np.poly1d(np.polyfit(X, y, 15))(X_curve)

# Plot
plt.scatter(X, y, color="black", label="Data")
plt.plot(X_curve, y_true, "g--", label="True function")
plt.plot(X_curve, y_overfit, "orange", label="Overfit")
plt.title("Overfitting (High Variance)")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
np.random.seed(0)
X = np.linspace(-3, 3, 30)
y = np.sin(X) + np.random.normal(0, 0.2, X.shape)

# Curve
X_curve = np.linspace(-3, 3, 200)
y_true = np.sin(X_curve)

# Good Fit (Polynomial degree 5)
y_goodfit = np.poly1d(np.polyfit(X, y, 5))(X_curve)

# Plot
plt.scatter(X, y, color="black", label="Data")
plt.plot(X_curve, y_true, "g--", label="True function")
plt.plot(X_curve, y_goodfit, "b", label="Good Fit")
plt.title("Good Fit (Balanced)")
plt.legend()
plt.show()
```
